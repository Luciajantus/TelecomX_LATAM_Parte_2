#📄Informe final

#### El objetivo de este proyecto fue identificar los factores que impulsan la evasión de clientes (Churn). Se analizó una base de datos de 7,043 clientes para detectar patrones de comportamiento y proponer soluciones estratégicas.

#### Se realizó un proceso de limpieza de la información:

Importación: Se cargaron datos desde un formato JSON anidado, normalizándolos en una estructura de tabla.

Corrección de Tipos: Transformé las columnas de texto a número (float64), manejando errores y rellenando vacíos con 0.

Tratamiento de Nulos: Identifiqué y eliminé 224 registros que tenían el campo de "Evasión" vacío, ya que sin ese dato el análisis perdería eficacia.

Estandarización: Convertí las respuestas categóricas (Yes/No) en valores binarios (1/0) usando funciones Lambda y el método .map().

Gráficas: Creé varias gráficas para visualizar el comportamiento del cliente y sus fugas en base a lo gastado y su duración en el tiempo.


#### Mediante visualizaciones, se descubrieron los patrones del cliente que se va de la empresa:

Distribución General: Aproximadamente el 26.5% de la base de clientes ha abandonado la compañía.

Existe una "Zona Crítica" en los primeros 12 meses. Los clientes nuevos tienen una tasa de fuga altísima comparada con los veteranos. Además, los clientes que pagan más de $70.00 mensuales (planes premium o con muchos adicionales) muestran una mayor tendencia a la evasión.

#### **Conclusiones:**
El tiempo del Contrato es Clave: Los clientes con contratos "Mes a mes" son los principales responsables de la evasión, ya que no tienen compromiso a largo plazo, facilitando la salida.

Precio: Los usuarios que pagan con Cheque Electrónico y tienen cargos mensuales altos son los más propensos a irse, posiblemente atraídos por ofertas de la competencia.

Fidelización: Una vez que un cliente supera los 24 meses de permanencia, su probabilidad de irse cae drásticamente.



#### **Recomendaciones**
Incentivar el Contrato Anual: Crear promociones para migrar a los clientes de "Mes a mes" a contratos de uno o dos años.

Plan de Bienvenida: Implementar un programa de atención especial y descuentos durante los primeros 6 meses de vida del cliente.

Revisión de las tarifas más caras: Analizar si los clientes que pagan más de $70 están recibiendo un valor percibido acorde o si se sienten sobrecargados, ofreciéndoles beneficios adicionales.

#### Creación de modelos

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, recall_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python
modelo_log = LogisticRegression(random_state=42)
modelo_log.fit(X_train_res, y_train_res)
y_pred_log = modelo_log.predict(X_test)
```

```python
modelo_rf = RandomForestClassifier(n_estimators=100, random_state=42)
modelo_rf.fit(X_train_res, y_train_res)
y_pred_rf = modelo_rf.predict(X_test)
```

```python
def evaluar(nombre, y_real, y_pred):
    acc = accuracy_score(y_real, y_pred)
    rec = recall_score(y_real, y_pred)
    print(f"--- {nombre} ---")
    print(f"Exactitud (Accuracy): {acc:.2%}")
    print(f"Exhaustividad (Recall): {rec:.2%}")
    print("-" * 20)

evaluar("Regresión Logística", y_test, y_pred_log)
evaluar("Random Forest", y_test, y_pred_rf)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 6))

ConfusionMatrixDisplay.from_predictions(y_test, y_pred_log,
                                        display_labels=['No Churn', 'Churn'],
                                        cmap='Blues', ax=ax[0])
ax[0].set_title('Matriz: Regresión Logística')

ConfusionMatrixDisplay.from_predictions(y_test, y_pred_rf,
                                        display_labels=['No Churn', 'Churn'],
                                        cmap='Greens', ax=ax[1])
ax[1].set_title('Matriz: Random Forest')

plt.tight_layout()
plt.show()
```

#### Evaluación de los Modelos


```python
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score

```

```python
def mostrar_metricas(modelo, X_train, y_train, X_test, y_test, nombre):

    y_pred_train = modelo.predict(X_train)
    y_pred_test = modelo.predict(X_test)

    print(f"--- MÉTRICAS: {nombre} ---")
    print(f"{'Métrica':<15} | {'Train':<10} | {'Test':<10}")
    print("-" * 40)

    metricas = [
        ("Exactitud", accuracy_score),
        ("Precisión", precision_score),
        ("Recall", recall_score),
        ("F1-score", f1_score)
    ]

    for nombre_m, func in metricas:
        m_train = func(y_train, y_pred_train)
        m_test = func(y_test, y_pred_test)
        print(f"{nombre_m:<15} | {m_train:<10.2%} | {m_test:<10.2%}")
    print("\n")


mostrar_metricas(modelo_log, X_train_res, y_train_res, X_test, y_test, "Regresión Logística")
mostrar_metricas(modelo_rf, X_train_res, y_train_res, X_test, y_test, "Random Forest")
```

#### Conclusión

**Objetivo:** Predecir la fuga de clientes (Churn) y determinar los factores clave que impulsan la cancelación para proponer acciones de mitigación.

**Pipeline de Datos**
Para asegurar que los modelos recibieran información de calidad, se realizaron los siguientes pasos:

Limpieza: Eliminación de duplicados y variables irrelevantes.

Codificación: Transformación de variables categóricas mediante One-Hot Encoding.

Escalado: Aplicar StandardScaler para normalizar variables con distintas magnitudes como tenure y Charges.

Balanceo (SMOTE): Debido al desbalance original (73% se quedan vs 26% se van), generé datos sintéticos para equilibrar el aprendizaje del modelo al 50/50

**Evaluación de Modelos Predictivos:**
Se comparó una Regresión Logística (basada en coeficientes) contra un Random Forest (basado en árboles).

1.   Exactitud (Accuracy)-Regresión Logística = 76.72%
2.   Exhaustividad (Recall) - Regresión Logística = **68.72%**
3.   Exactitud (Accuracy)-Random Forest= **78.35%**
4.   Exhaustividad (Recall)-Random Forest= 52.94%






- El Random Forest muestra señales de overfitting (mejor precisión en entrenamiento que en prueba) y un Recall bajo.

- La Regresión Logística es la ganadora para este caso de negocio. Su Recall del 68.7% permite identificar a la gran mayoría de los clientes en riesgo, minimizando los "falsos negativos" que tanto le cuestan a la empresa.

**Hallazgos y Plan de Acción**

Tras analizar la importancia de las variables, estos son los hallazgos principales:

1. Los puntos difíciles:
- Contratos Mensuales: Son el mayor predictor de fuga. La falta de compromiso a largo plazo facilita la salida.

- Fibra Óptica: Presenta una correlación positiva con el Churn, sugiriendo posibles problemas de precio o estabilidad en este servicio específico.

- Los primeros 12 meses son críticos. Superado el año, la probabilidad de fuga cae drásticamente.



2. Estrategias de Retención
- Campañas de beneficios exclusivos para clientes con menos de 6 meses de antigüedad.

- Incentivar el paso de "Mes a mes" a contratos anuales mediante descuentos en los cargos mensuales.

- Auditoría de calidad y precio en el servicio de fibra óptica para alinear las expectativas del cliente con el costo.

**Conclusión**
El modelo de Regresión Logística entrenado permite actuar de forma preventiva sobre casi el 70% de los clientes potenciales a irse. Con las estrategias propuestas, Telecom X puede pasar de una postura reactiva a una proactiva, protegiendo los ingresos y mejorando la lealtad del cliente.
